### Frequentist method of estimating $\mu$ 

Prior to using Bayesian methods, first consider the frequentist method for estimating model parameters, in this case the estimation of the mean of a Poisson distribution. The function `poisson_logprob()` returns the likelihood of a set of data `y_obs` given a particular Poisson model and associated parameter. Then we use `opt.minimize_scalar()` to find the optimal value of $\mu$ to maximise the likelihood.

```python
y_obs = messages['time_delay_seconds'].values

def poisson_logprob(mu, sign=-1):
    return np.sum(sign*stats.poisson.logpmf(y_obs, mu=mu))

freq_results = opt.minimize_scalar(poisson_logprob)
%time print("The estimated value of mu is: %s" % freq_results['x'])
```

Under the hood, `y_obs` is a list of the time delays to respond to each of the 338 messages:

For a particular $\mu$, the probability mass function looks like the bar plots above, whilst the negative of the `logpmf` function looks like:

```python
mu_example = [15,25,freq_results.x, 40]

fig,ax= plt.subplots(2,1)
# ax[0].bar(y_obs, stats.poisson.pmf(y_obs,mu_example[0]), label="$\mu$ = 15",alpha = 0.2)
# ax[0].bar(y_obs, stats.poisson.pmf(y_obs,mu_example[1]),label="$\mu$ = 25",alpha = 0.2)
ax[0].bar(y_obs, stats.poisson.pmf(y_obs,mu_example[2]),label=f"$\mu$ = {freq_results.x:0.3f}",alpha = 0.2)
# ax[0].bar(y_obs, stats.poisson.pmf(y_obs,mu_example[3]),label=f"$\mu$ = {mu_example[3]}", alpha = 0.1)
ax[0].set(ylabel="PMF")
# ax[1].bar(y_obs, -stats.poisson.logpmf(y_obs,mu_example[0]),label="$\mu$ = 15",alpha = 0.2)
# ax[1].bar(y_obs, -stats.poisson.logpmf(y_obs,mu_example[1]),label="$\mu$ = 25",alpha = 0.2)
ax[1].bar(y_obs, stats.poisson.logpmf(y_obs,mu_example[2]),label=f"$\mu$ = {freq_results.x:0.3f}", alpha = 0.2)
# ax[1].bar(y_obs, -stats.poisson.logpmf(y_obs,mu_example[3]),label="$\mu$ = 40", alpha = 0.1)
ax[1].set(ylabel="log(PMF)")
ax[0].set(ylim=[0,0.15])
# ax[1].set(ylim=[0,60], xlabel="response time (seconds)")
ax[0].legend()
ax[1].legend()

plt.show()
```

The saturation of each bar corresponds to the frequency that particular message response time ($Y_i$) occured in the data. When we perform the sum, two bars of the same height, but different saturations, contribute proportionaly differing amounts. If most of the saturation occurs near the minumum in the `logpmf`, then the summation will be minimised.
When a Poisson distribution with the "correct" mean $\mu$ is fit to the data, the sum will be smallest, and an optimisation routine can find this $\mu$ to minimise the sum.

```python
y = [1,1,1,1,1,1,1,1,1,2, 2, 3, 4, 5, 6, 2,2,2,2,3,3,3,3]
fig,ax = plt.subplots()
# test_mean = 5
test_mean = np.mean(y)
ax.bar(y, stats.poisson.pmf(y, mu=test_mean), alpha = 0.2)
plt.show()
print(f"mean = {np.mean(y):0.5f}, sum={np.sum(stats.poisson.pmf(y, mu=test_mean)):0.5f}")
```

```python
with pm.Model() as model:
    mu = pm.Uniform('mu', lower=0, upper=60)
    likelihood = pm.Poisson('likelihood', mu=mu, observed=messages['time_delay_seconds'].values)

    start = pm.find_MAP()
    step = pm.Metropolis()
    trace = pm.sample(200000, step=step, initvals=start, progressbar=True)
```

```python
_ = pm.plot_trace(trace, var_names='mu', lines={'mu': freq_results['x']})
```

```python
posterior = trace.posterior.stack(sample=['chain', 'draw'])
vals = posterior['mu']
fig = plt.figure(figsize=(11,3))
plt.subplot(121)
_ = plt.title('Burnin trace')
_ = plt.ylim(ymin=16.5, ymax=19.5)
_ = plt.plot(vals[:1000]) 
fig = plt.subplot(122)
_ = plt.title('Full trace')
_ = plt.ylim(ymin=16.5, ymax=19.5)
_ = plt.plot(vals)
```

```python
x_lim = 60
burnin = 50000

y_pred = trace[burnin:].get_values('y_pred')
mu_mean = trace[burnin:].get_values('mu').mean()

fig = plt.figure(figsize=(10,6))
fig.add_subplot(211)

_ = plt.hist(y_pred, range=[0, x_lim], bins=x_lim, histtype='stepfilled', color=colors[1])   
_ = plt.xlim(1, x_lim)
_ = plt.ylabel('Frequency')
_ = plt.title('Posterior predictive distribution')

fig.add_subplot(212)

_ = plt.hist(messages['time_delay_seconds'].values, range=[0, x_lim], bins=x_lim, histtype='stepfilled')
_ = plt.xlabel('Response time in seconds')
_ = plt.ylabel('Frequency')
_ = plt.title('Distribution of observed data')

plt.tight_layout()
```
